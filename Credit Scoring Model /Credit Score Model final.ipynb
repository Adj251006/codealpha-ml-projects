```python
df.columns = [
    "duration", "credit_amount", "installment_rate", "residence_since",
    "age", "existing_credits", "num_dependents",
    "checking_account", "credit_history", "purpose", "savings",
    "employment", "personal_status", "other_debtors",
    "property", "other_installment_plans", "housing",
    "job", "telephone", "foreign_worker", "credit_risk"
]
```

```python
X = df.drop('credit_risk', axis=1)
Y = df['credit_risk']
```

```python
from sklearn.model_selection import train_test_split

Xtrain,Xtest,Ytrain,Ytest = train_test_split(X, Y, test_size=0.2, random_state=42, stratify=Y)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
Xtrain_scaled = scaler.fit_transform(Xtrain)
Xtest_scaled  = scaler.transform(Xtest)
```

```python
from sklearn.linear_model import LogisticRegression

logistic_tuned = LogisticRegression(C=0.5, max_iter=1000, solver='liblinear')
logistic_tuned.fit(Xtrain_scaled,Ytrain)
```

```python
y_pred = logistic_tuned.predict(Xtest_scaled)
```

```python
from sklearn.metrics import classification_report
print(classification_report(Ytest,y_pred))
```

```python
#0 → Not creditworthy (bad customer)

#1 → Creditworthy (good customer)
```

```python
from sklearn.metrics import accuracy_score

# predict on test data
y_pred = logistic_tuned.predict(Xtest_scaled)

# accuracy
accuracy = accuracy_score(Ytest, y_pred)
print("Model Accuracy:", accuracy)
```

```python
creditworthy_user = Xtrain[Ytrain == 1].iloc[0].values.reshape(1, -1)
creditworthy_user_scaled = scaler.transform(creditworthy_user)

pred = logistic_tuned.predict(creditworthy_user_scaled)
prob = logistic_tuned.predict_proba(creditworthy_user_scaled)[0][1]

print("Prediction:", pred)
print("Probability:", prob)
```

```python
if pred[0] == 1:
    print("✅ User is CREDITWORTHY")
else:
    print("❌ User is NOT CREDITWORTHY")
```
